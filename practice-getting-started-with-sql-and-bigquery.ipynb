# BigQuery commands

```python
#import the python package for BigQuery
from google.cloud import bigquery
```

```python
#create bigquery client
client = bigquery.Client()
```

Dataset in use is : [Hacker News](http://https://news.ycombinator.com/)

In BigQuery, each dataset is contained in a corresponding project. In this case, our hacker_news dataset is contained in the bigquery-public-data project.

```python
#Reference to hacker news dataset.
dataset_ref = client.dataset("hacker_news", project="bigquery-public-data")

#fecth the dataset
dataset = client.get_dataset(dataset_ref)
```

Every dataset is just a collection of tables. You can think of a dataset as a spreadsheet file containing multiple tables, all composed of rows and columns.

We use the **list_tables()** method to list the tables in the dataset.

```python
#list all the tables in "hacker_news" dataset
tables = list(client.list_tables(dataset))

for table in tables:
    print(table.table_id)
```

Fetch the "full" table in the "hacker_news" dataset.

```python
# ref to full table
table_ref = dataset_ref.table("full")

#fetch the table
table = client.get_table(table_ref)
```

# Table schema
The structure of a table is called its schema. Let's check the *full* table schema.

```python
#print information of all the columns in the table.
table.schema
```

```python
# check out first 5 rows of "full" table
rows = client.list_rows(table, max_results=5)
rows
```

```python
rows.to_dataframe()
```

```python
# preview first five entries in the "by" column:

client.list_rows(table,selected_fields=table.schema[4:5], max_results=5).to_dataframe()
```
